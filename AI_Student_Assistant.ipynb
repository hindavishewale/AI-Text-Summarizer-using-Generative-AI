```python
from transformers import pipeline
import gradio as gr
```

```python
chatbot = pipeline("text-generation", model="gpt2")
summarizer = pipeline("summarization")
classifier = pipeline("sentiment-analysis")
```

```python
def student_ai(task, text):

    if task == "Chatbot":
        result = chatbot(text, max_length=100)
        return result[0]['generated_text']

    elif task == "Summarize Notes":
        result = summarizer(text, max_length=60, min_length=20)
        return result[0]['summary_text']

    elif task == "Resume Screening":
        result = classifier(text)
        return result[0]['label']
```

```python
interface = gr.Interface(
    fn=student_ai,
    inputs=[
        gr.Dropdown(["Chatbot","Summarize Notes","Resume Screening"]),
        gr.Textbox(lines=6)
    ],
    outputs="text",
    title="AI Student Assistant"
)

interface.launch()
```
